## Use cv2's Contrast Limited Adaptive Histogram Equalization to rescale the L color space contrast

```python
def col_equalise(img):
    lab= cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
    # split by channel
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl = clahe.apply(l)
    limg = cv2.merge((cl,a,b))
    final = cv2.cvtColor(limg, cv2.COLOR_LAB2RGB)
    return final
```

```python
for img in os.listdir(train_dir):
    image = os.path.join(train_dir,img)
    image = cv2.imread(str(image))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = col_equalise(image)
    plt.imsave("D:/Capstone_Project/Train/"+str(img), image)
```

## Show that it works when loading tf.tensor for adaptive histogram

```python
init = tf.initialize_all_variables()    
sess = tf.Session()
sess.run(init)

result_dir = 'C:/Users/n3rDx/Desktop/Homework Upload/Capstone/results/'

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(25, 25),
                         sharex=True, sharey=True)
ax = axes.ravel()

for img in os.listdir(result_dir):
    image = tf.io.read_file(os.path.join(result_dir,img))
    image = tf.image.decode_jpeg(image)
    image = tf.cast(image, tf.float32)
    X_train = tf.image.grayscale_to_rgb(tf.image.rgb_to_grayscale(image))
    ax[1].imshow(sess.run(X_train/256))
    ax[1].title.set_text("L adjusted grayscale picture")
    ax[2].imshow(sess.run(image/255))
    ax[2].title.set_text("L adjusted colored picture")
    break
    
for img in os.listdir(train_dir):
    Oimage = os.path.join(train_dir,img)
    Oimage = cv2.imread(str(Oimage))
    Oimage = cv2.cvtColor(Oimage, cv2.COLOR_BGR2RGB)
    ax[0].imshow(Oimage)
    ax[0].title.set_text("Original picture")
    break
    
Oimage = tf.cast(Oimage,tf.float32)
print(sess.run(tf.math.reduce_mean(tf.image.ssim_multiscale(X_train,image,max_val=2e-4))))
```
